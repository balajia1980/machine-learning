## Label Encoding using Binary Encoder for Categorical Variable

```python
vals_to_replace_sex = {'Male':'1', 'Female':'0'}
vals_to_replace_income = {'>50K':'1', '<=50K':'0'}
data['income']=data['income'].map(vals_to_replace_income).astype(str).astype(int)
data['sex'] = data['sex'].map(vals_to_replace_sex).astype(str).astype(int)
encoder_reln = ce.BinaryEncoder(cols=['relationship'])
df_binary_relationship = encoder_reln.fit_transform(data['relationship'])
encoder_race = ce.BinaryEncoder(cols=['race'])
df_binary_race = encoder_race.fit_transform(data['race'])
encoder_work = ce.BinaryEncoder(cols=['workclass'])
df_binary_work = encoder_work.fit_transform(data['workclass'])
encoder_marital = ce.BinaryEncoder(cols=['marital.status'])
df_binary_marital = encoder_marital.fit_transform(data['marital.status'])
encoder_native = ce.BinaryEncoder(cols=['native.country'])
df_native = encoder_native.fit_transform(data['native.country'])
encoder_occupn = ce.BinaryEncoder(cols=['occupation'])
df_occupn = encoder_occupn.fit_transform(data['occupation'])
```

```python
data=[data, df_binary_marital, df_binary_relationship, df_binary_race, df_binary_work, df_native,df_occupn]
data=pd.concat(data,axis=1)
data.shape
```

```python
data.info()
```

```python
cols = [0,2,4,6,7,8,9,14]
data.drop(data.columns[cols],axis=1,inplace=True)
X=data.drop(['income'], axis=1)
y=data['income']
X.shape,y.shape
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=5)
```

## Build ML model and check the accuracy score

```python
def run_randomForest(X_train,X_test,y_train,y_test):
    reg = RandomForestClassifier(n_estimators=100, random_state=0, n_jobs=-1)
    reg.fit(X_train, y_train)
    y_pred=reg.predict(X_test)
    print("Accuracy Score: ")
    print(accuracy_score(y_test,y_pred))
```

```python
run_randomForest(X_train, X_test, y_train, y_test)
```
